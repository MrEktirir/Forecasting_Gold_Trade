```python
# IQR yöntemi ile outlier tespiti ve temizleme
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# 'Qty' ve 'Fobvalue' sütunlarındaki outlier'ları temizleyelim
df = remove_outliers(df, 'Qty')
df = remove_outliers(df, 'Fobvalue')
```

```python
from sklearn.linear_model import LinearRegression

ghana_grouped = myth_1_ghana.groupby(['Category', 'RefYear']).agg({'Fobvalue': 'sum'}).reset_index()
nigeria_grouped = myth_1_nigeria.groupby(['Category', 'RefYear']).agg({'Fobvalue': 'sum'}).reset_index()
```

```python
from sklearn.linear_model import LinearRegression

# Bölge bazında yıllık toplam Qty değerleri için boş sözlükler oluştur
ghana_grouped = myth_1_ghana.groupby(['Category', 'RefYear']).agg({'Qty': 'sum'}).reset_index()
nigeria_grouped = myth_1_nigeria.groupby(['Category', 'RefYear']).agg({'Qty': 'sum'}).reset_index()

# Her bir bölge için doğrusal model kurma ve tahmin yapma fonksiyonu
def fit_predict_region(data, future_years):
    categories = data['Category'].unique()  # Bölgeleri al
    predictions = {}  # Tahminler için boş sözlük

    for c in categories:
        # Belirli bir bölgeyi filtrele
        region_data = data[data['Category'] == c]
        
        # Doğrusal regresyon modeli
        model = LinearRegression()
        X = region_data['RefYear'].values.reshape(-1, 1)  # Yıllar (bağımsız değişken), 2D array haline getirildi
        y = region_data['Qty']  # Qty değerleri (bağımlı değişken)
        
        # Modeli eğit
        model.fit(X, y)
        
        # Gelecekteki yıllar için tahmin
        future_X = np.array(future_years).reshape(-1, 1)  # Tahmin yılları da 2D array haline getirildi
        predictions[c] = model.predict(future_X).tolist()

    return predictions

# Gana ve Nijerya için tahminler
ghana_predictions = fit_predict_region(ghana_grouped, [2024, 2025, 2026])
nigeria_predictions = fit_predict_region(nigeria_grouped, [2024, 2025, 2026])
```

```python
def plot_predictions(data, predictions, title):
    # Gerçek verilerin ve tahminlerin birleştirilmesi
    years = list(data['RefYear'].unique()) + [2024, 2025, 2026]
    categories = data['Category'].unique()
    
    # Grafik çizimi için figür oluşturma
    fig, axes = plt.subplots(nrows=len(categories), ncols=1, figsize=(10, 5 * len(categories)), sharex=True)
    
    if len(categories) == 1:
        axes = [axes]  # Eğer tek grafik varsa, döngüyü düzgün çalıştırmak için listeye çevir
    
    for ax, c in zip(axes, categories):
        # Bölgeye göre verileri filtrele
        region_data = data[data['Category'] == c]
        region_preds = predictions[c]
        
        # Gerçek veriler
        sns.lineplot(x='RefYear', y='Qty', data=region_data, ax=ax, label='Actual', marker='o')
        
        # Tahminler
        ax.plot([2024, 2025, 2026], region_preds, 'r--', label='Predicted')
        
        ax.set(title=f'{title} - {c}', ylabel='Qty')
        ax.legend()
    
    plt.tight_layout()
    plt.show()

# Gana ve Nijerya için grafik çizimi
plot_predictions(ghana_grouped, ghana_predictions, "Ghana FOB Value Predictions by Region")
plot_predictions(nigeria_grouped, nigeria_predictions, "Nigeria FOB Value Predictions by Region")
```

```python
def combine_data_predictions(data, predictions):
    future_years = [2024, 2025, 2026]
    prediction_rows = []
    for category, pred_values in predictions.items():
        for year, value in zip(future_years, pred_values):
            prediction_rows.append({'RefYear': year, 'Category': category, 'Qty': value, 'Type': 'Predicted'})
    
    df_predictions = pd.DataFrame(prediction_rows)
    
    data['Type'] = 'Actual'
    
    combined_df = pd.concat([data, df_predictions], ignore_index=True)
    
    combined_df.sort_values(by=['Category', 'RefYear'], inplace=True)
    
    return combined_df

combined_ghana = combine_data_predictions(ghana_grouped, ghana_predictions)
combined_nigeria = combine_data_predictions(nigeria_grouped, nigeria_predictions)
```
